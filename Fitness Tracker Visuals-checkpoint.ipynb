# Coding Same Visuals Using Different Libraries

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pygal
```

```python
df = pd.read_csv("Fitness_trackers_updated.csv")
```

```python
df.head()
```

```python
df['Selling Price'] = df['Selling Price'].replace(',','', regex=True)
df['Original Price'] = df['Original Price'].replace(',','', regex=True)
```

```python
df.head()
```

```python
df['Selling Price'] = df['Selling Price'].astype(float)
df['Original Price'] = df['Original Price'].astype(float)
```

```python
df.dtypes
```

## Chart 1: Horizontal Bar

```python
df_plt = df.groupby("Brand Name")["Rating (Out of 5)"].mean().sort_values(ascending=True)
```

```python
df_plt.head(25)
```

## Pandas

```python
bar_plot = df_plt.plot(kind='barh', 
                        title= 'AVG Rating Fitness Tracker',
                        ylabel='AVG Rating',
                        xlabel='Brand',
                        figsize=(6, 5)
                       )

bar_plot.bar_label(bar_plot.containers[0])
plt.xlim([0,5.5])
plt.show()
```

## Matplotlib

```python
br_plot = df_plt.plot.barh(color= "dodgerblue")

plt.title("AVG Rating Ftiness Tracker")
br_plot.bar_label(br_plot.containers[0])
plt.xlim([0,5.5])
plt.show()
```

## Seaborn

```python
df_seaborn = df.groupby(['Brand Name'])['Rating (Out of 5)'].mean().reset_index().sort_values('Rating (Out of 5)', ascending=False)
```

```python
# create plot
b_c = sns.barplot(x = 'Rating (Out of 5)', 
            y = 'Brand Name', 
            data = df_seaborn,
            palette = 'GnBu_d',  
            capsize = 0.05,             
            saturation = 8,
            errcolor = 'gray', 
            errwidth = 1   
            )

plt.title("AVG Rating Ftiness Tracker")
b_c.bar_label(b_c.containers[0])
plt.xlim([0,5.5])
plt.show()
```

## Pygal

```python
df_plt2 = df.groupby("Brand Name")["Rating (Out of 5)"].mean().sort_values(ascending=False)
```

```python
from IPython.display import display, HTML

base_html = """
<!DOCTYPE html>
<html>
  <head>
  <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/svg.jquery.js"></script>
  <script type="text/javascript" src="https://kozea.github.io/pygal.js/2.0.x/pygal-tooltips.min.js""></script>
  </head>
  <body>
    <figure>
      {rendered_chart}
    </figure>
  </body>
</html>
"""
```

```python
bar_chart = pygal.HorizontalBar(height=580)
bar_chart.title = 'AVG Rating Ftiness Tracker'
[bar_chart.add(x[0], x[1]) for x in df_plt2.items()]
display(HTML(base_html.format(rendered_chart=bar_chart.render(is_unicode=True))))
```
