Libraries used during this exercise:
Matplotlib
Seaborn
Plotly
Pygal

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pygal
```

```python
df = pd.read_csv("Fitness_trackers_updated.csv")
```

```python
df.head()
```

```python
df['Selling Price'] = df['Selling Price'].replace(',','', regex=True)
df['Original Price'] = df['Original Price'].replace(',','', regex=True)
```

```python
df.head()
```

```python
df['Selling Price'] = df['Selling Price'].astype(float)
df['Original Price'] = df['Original Price'].astype(float)
```

```python
df.dtypes
```

```python
df_plt = df.groupby("Brand Name")["Rating (Out of 5)"].mean().sort_values()
```

```python
df_plt.head()
```

```python
df_plt.plot(kind='barh', title=' AVG Rating Fitness Tracker', ylabel='AVG Rating',
         xlabel='Brand', figsize=(6, 5))
```

```python
df_plt.plot.barh(color= "dodgerblue")
plt.show()
```

```python
df_sns = df.groupby(['Brand Name']).agg(mean_rating=("Rating (Out of 5)", 'mean'))
df_sns = df.reset_index()
```

```python
sns.barplot(data=df_sns,
            ci=None,
            x="Rating (Out of 5)", 
            y="Brand Name")
plt.xticks(rotation=90)
plt.show()
```
